```python
import pandas as pd
```

```python
df = pd.read_csv('oe_relevant.csv', low_memory=False)
```

```python
df
```

```python
df.info()
```

```python
df.tail(50)
```

```python
# Filter out gameids that don't have exactly 10 entries
df_filtered = df.groupby('gameid').filter(lambda x: len(x) == 10)
```

```python
df_filtered.info()
```

```python
# Sort each game by 'participantid'
df_sorted = df_filtered.sort_values(by=['gameid', 'participantid'])

df_sorted.head(50)
```

```python
df_sorted.reset_index(drop=True, inplace=True)
df_sorted.head(20)
```

```python
df_sorted.info()
```

```python
df_sorted.to_csv('cleaned.csv', index=False)
```
