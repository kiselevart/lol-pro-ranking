```python
import pandas as pd
```

```python
df = pd.read_csv('df_full_lper.csv')
df
```

```python
league_strengths = {
    'LEC': 1542,
    'LCS': 1486,
    'LCK': 1873
}
```

```python
df['Adj_LPER'] = df.apply(lambda row: round(row['LPER'] * league_strengths[row['league']]), axis=1)
df
```

```python
df_avg_lper = df.groupby('playername')['Adj_LPER'].mean().reset_index()
df_avg_lper['Adj_LPER'] = df_avg_lper['Adj_LPER'].round().astype(int)
df_avg_lper.rename(columns={'Adj_LPER': 'LPER'}, inplace=True)
df_avg_lper
```

```python
df_lper = df_avg_lper.sort_values(by='LPER', ascending=False)
df_lper
```

```python
df_lper.head(50)
```

```python
unique_players_by_role = df.groupby('position')['playername'].unique().reset_index()
unique_players_by_role
```

```python
bot_unique = unique_players_by_role[unique_players_by_role['position'] == 'bot']['playername'].values[0]
jng_unique = unique_players_by_role[unique_players_by_role['position'] == 'jng']['playername'].values[0]
mid_unique = unique_players_by_role[unique_players_by_role['position'] == 'mid']['playername'].values[0]
sup_unique = unique_players_by_role[unique_players_by_role['position'] == 'sup']['playername'].values[0]
top_unique = unique_players_by_role[unique_players_by_role['position'] == 'top']['playername'].values[0]
```

```python
bot_LPER = df_lper[df_lper['playername'].isin(bot_unique)].sort_values(by='LPER', ascending=False).reset_index(drop=True)
bot_LPER.index += 1
jng_LPER = df_lper[df_lper['playername'].isin(jng_unique)].sort_values(by='LPER', ascending=False).reset_index(drop=True)
jng_LPER.index += 1
mid_LPER = df_lper[df_lper['playername'].isin(mid_unique)].sort_values(by='LPER', ascending=False).reset_index(drop=True)
mid_LPER.index += 1
sup_LPER = df_lper[df_lper['playername'].isin(sup_unique)].sort_values(by='LPER', ascending=False).reset_index(drop=True)
sup_LPER.index += 1
top_LPER = df_lper[df_lper['playername'].isin(top_unique)].sort_values(by='LPER', ascending=False).reset_index(drop=True)
top_LPER.index += 1
```

```python
tables = [top_LPER, jng_LPER, mid_LPER, bot_LPER, sup_LPER]
tables
```

```python
# Function to add difference columns
def add_differences(table):
    top_lper = table['LPER'].max()
    avg_lper = table['LPER'].mean()
    
    table['Difference_From_Top'] = table['LPER'] - top_lper
    table['Difference_From_Average'] = table['LPER'] - avg_lper
    return table

# Apply the function to all tables
tables = [add_differences(table) for table in tables]

# Display the updated tables
for i, table in enumerate(tables):
    print(f"Table {i + 1}:\n{table}\n")
```

```python
# Group by playername and position, count games, and create a new column for the main role
player_roles = (
    df.groupby(['playername', 'position'])
    .size()
    .unstack(fill_value=0)  # Make positions into columns
    .reset_index()          # Ensure playername is a regular column
)

# Remove the index name ('position') if it persists
player_roles.columns.name = None

# Add a new 'position' column with the role that has the highest game count
player_roles['position'] = player_roles.iloc[:, 1:].idxmax(axis=1)

# Display the updated table
player_roles.head(50)
```

```python
# Ensure consistent case in player_roles['position']
player_roles['position'] = player_roles['position'].str.lower()

# Initialize the dictionary to store filtered tables
filtered_tables = {}

# Define table positions
table_positions = ['top', 'jng', 'mid', 'bot', 'sup']

# Loop through indices instead of using zip
for i in range(len(tables)):
    table = tables[i]
    position = table_positions[i]
    
    # Normalize case and trim whitespace in player names for consistency
    table['playername'] = table['playername'].str.strip().str.lower()
    player_roles['playername'] = player_roles['playername'].str.strip().str.lower()

    # Filter players whose main position matches the current table's position
    filtered_tables[position] = table[
        table['playername'].isin(
            player_roles[player_roles['position'] == position]['playername']
        )
    ]

filtered_tables
```

```python
top_LPER = filtered_tables['top'].reset_index(drop=True)
jng_LPER = filtered_tables['jng'].reset_index(drop=True)
mid_LPER = filtered_tables['mid'].reset_index(drop=True)
bot_LPER = filtered_tables['bot'].reset_index(drop=True)
sup_LPER = filtered_tables['sup'].reset_index(drop=True)
```

```python
# Define a function to normalize the 'LPER' values within each table using Min-Max Scaling
def normalize_lper(table):
    # Min-Max normalization: (x - min) / (max - min)
    table['LPER_normalized'] = (table['LPER'] - table['LPER'].min()) / (table['LPER'].max() - table['LPER'].min())
    return table

# Normalize 'LPER' in each table
top_LPER = normalize_lper(top_LPER)
jng_LPER = normalize_lper(jng_LPER)
mid_LPER = normalize_lper(mid_LPER)
bot_LPER = normalize_lper(bot_LPER)
sup_LPER = normalize_lper(sup_LPER)

# Merge all filtered tables together
merged_LPER = pd.concat([top_LPER, jng_LPER, mid_LPER, bot_LPER, sup_LPER], ignore_index=True)

# Sort by 'Difference_From_Average' in descending order
merged_LPER = merged_LPER.sort_values(by='LPER_normalized', ascending=False)

# Reset index for the final DataFrame
merged_LPER = merged_LPER.reset_index(drop=True)

merged_LPER.head(50)
```

```python
# Function to calculate mean, std, and Z-score for each position
def calculate_z_score(table):
    position_mean = table['LPER'].mean()  # Mean for each role
    position_std = table['LPER'].std()    # Standard deviation for each role
    
    # Calculate the Z-score (number of standard deviations away from the mean)
    table['Z-score'] = (table['LPER'] - position_mean) / position_std
    
    # Add the mean and standard deviation as columns for reference
    table['Role_Mean'] = position_mean
    table['Role_Std'] = position_std
    
    return table

# Apply to each table
top_LPER = calculate_z_score(top_LPER)
jng_LPER = calculate_z_score(jng_LPER)
mid_LPER = calculate_z_score(mid_LPER)
bot_LPER = calculate_z_score(bot_LPER)
sup_LPER = calculate_z_score(sup_LPER)

# Merge all tables together
merged_LPER = pd.concat([top_LPER, jng_LPER, mid_LPER, bot_LPER, sup_LPER], ignore_index=True)

# Sort by Z-score (highest absolute deviation)
merged_LPER = merged_LPER.sort_values(by='Z-score', ascending=False)

# Reset index for the final DataFrame
merged_LPER = merged_LPER.reset_index(drop=True)

merged_LPER.head(50)
```

```python
df_z_score = merged_LPER.reset_index()[['playername', 'Z-score']]
df_z_score
```

```python
merged_LPER['Z-Elo'] = (merged_LPER['Z-score'] * 1000).round().astype(int)
merged_LPER
```

```python
# Calculate the minimum Z-Elo value
min_z_elo = merged_LPER['Z-Elo'].min()

# Add the absolute value of the minimum Z-Elo to all Z-Elo values to make them positive
merged_LPER['Z-Elo_Positive'] = merged_LPER['Z-Elo'] + abs(min_z_elo)

# Display the updated DataFrame
merged_LPER
```
