```python
import pandas as pd
```

```python
df = pd.read_csv('OEFULL.csv', low_memory=False)
```

```python
df_c = df[df['datacompleteness'] == 'complete']
df_c.drop(columns=['index', 'datacompleteness'], inplace=True)
df_c
```

```python
threshold = len(df_c) * 0.7
columns_to_drop = [col for col in df_c.columns if df_c[col].isnull().sum() > threshold and col not in ['playername', 'playerid']]
df_c = df_c.drop(columns=columns_to_drop)
df_c = df_c.dropna(thresh=threshold, axis=1)
df_c
```

```python
df_c.info()
for col in df_c.columns:
    print(col)
```

```python
columns_with_at = [col for col in df_c.columns if 'at' in col]
columns_with_at
```

```python
columns_with_at.remove('date')
columns_with_at.remove('patch')
columns_with_at.remove('deaths')
columns_with_at.remove('teamdeaths')
columns_with_at.remove('damagemitigatedperminute')
columns_with_at
```

```python
df_cd = df_c.drop(columns=columns_with_at)
df_cd
```

```python
df_cdg = df_cd[~df_cd['participantid'].isin([100, 200])]
df_cdg
```

```python
df_cdg.info()
```

```python
df_cdg = df_cdg.drop(columns=['year','split','date','patch','teamname','teamid','champion','ban1','ban2','ban3','ban4','ban5','doublekills','triplekills','quadrakills','pentakills','firstblood','firstbloodkill','firstbloodassist','firstbloodvictim'])
df_cdg
```

```python
df_cdg.drop(columns=['playerid'])
df_cdg
```

```python
df_cdg.info()
```

```python
numeric = df_cdg.select_dtypes(include=['float64', 'int64'])
numeric
```

```python
corr_matrix = numeric.corr()
corr_matrix.abs()
```

```python
corr_matrix = corr_matrix.abs()
corr_matrix
```

```python
# Threshold for high correlation
threshold = 0.9

# Find columns that are highly correlated
high_corr_pairs = [
    (col1, col2)
    for col1 in corr_matrix.columns
    for col2 in corr_matrix.columns
    if col1 != col2 and corr_matrix.loc[col1, col2] > threshold
]

print("Highly correlated feature pairs:")
print(high_corr_pairs)
```

```python
# Find columns to drop
to_drop = set()

for col1, col2 in high_corr_pairs:
    if col1 not in to_drop and col2 not in to_drop:
        to_drop.add(col2)  # Arbitrarily drop col2; you can apply custom logic here

print("Features to drop:")
print(to_drop)

# Drop redundant features
df_reduced = df_cdg.drop(columns=to_drop)
```

```python
df_notna = df_reduced.dropna()
df_notna
```

```python
df_notna = df_notna.drop(columns=['playerid'])
df_notna
```
